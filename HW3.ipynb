```python
#B4
h = int(input())
for i in range(1,h+1):
    if i==1:
        print(' '*(h-1) + '*' + ' '*(h-1))
    elif i==h:
        print('*'*(2*i-1))
    else:
        print(' '*(h-i) + '*' + ' '*(i*2-3) + '*' + ' '*(h-i))
```

```python
#B14
while True:
    mx = 0
    mn = 1000000000
    a = int(input())
    if a == -1:
        break
    mx = max(mx, a)
    mn = min(mn, a)
print("Max:", mx)
print("Min:", mn)
```

```python
#B16
n = int(input())
o,e = 0,0
while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        e += 1
    else:
        o += 1
    n //= 10
print("Even digits:", e)
print("Odd digits:", o)
```

```python
#B17
n = int(input())
digit_sum = 0
while n > 0:
    digit = n % 10
    digit_sum += digit
    n //= 10
print(digit_sum)
```

```python
#B19
A = int(input())
a,b,c = 0,1,1
while c < A:
    c = a + b
    a = b
    b = c
    if c > A:
        print(b)
        break
```
