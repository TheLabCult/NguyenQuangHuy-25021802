```python
#B5
a,b = map(int, input().split())
if a==b:
    print("a bằng b")
else:
    print(max(a,b))
```

```python
#B6
c = input()
if ord(c)>=48 and ord(c)<=57:
    print("c là chữ số")
elif ord(c)>=65 and ord(c)<=90 or ord(c)>=97 and ord(c)<=122:
    print("c là chữ cái")
```
