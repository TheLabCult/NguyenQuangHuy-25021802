```python
#W6A9
dict1, dict2 = {}, {}
a = input().split()
for i in a:
    i = i.split(':')
    dict1[i[0]] = int(i[1])
b = input().split()
for j in b:
    j = j.split(':')
    dict2[j[0]] = int(j[1])
dict3 = {}
for key in dict1:
    if key in dict2:
        dict3[key] = dict1[key] + dict2[key]
    else:
        dict3[key] = dict1[key]
for key in dict2:
    if key not in dict1:
        dict3[key] = dict2[key]
dict3.sort()
print(dict3)
```

```python
#W6A10
a = [int(i) for i in input().split()]
k = int(input())
tup = []
for i in range(len(a)):
    for j in range(i+1, len(a)):
        if a[i] + a[j] == k:
            tup.append((a[i], a[j]))
print(tup)
```

```python
#W6A12
a = (int(i) for i in input().split())
s = set(a)
mx = []
for i in s:
    mx.append([a.count(i), i])
mx.sort()
ptr = len(mx)-1
for i in range(len(mx)-1, 0, -1):
    if mx[i][0] == mx[i-1][0]:
        ptr -= 1
    else:
        break
print(mx[ptr][1])
```
