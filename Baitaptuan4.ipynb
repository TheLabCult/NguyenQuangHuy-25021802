```python
#W4A8
n = int(input())
def reverse(n):
    rev = 0
    while n > 0:
        dig = n % 10
        rev = rev * 10 + dig
        n //= 10
    return rev
cnt = 0
while reverse(n) != n:
    n += reverse(n)
    cnt += 1
print(cnt)
print(n)
```
