```python
#B3
a,b,c = map(int, input().split())
if a+b<=c or a+c<=b or b+c<=a:
    print("Khong la ba canh cua mot tam giac")
elif a==b and b==c:
    print("Tam giác đều")
elif a==b or a==c or b==c:
    print("Tam giác cân")
elif a*a+b*b==c*c or a*a+c*c==b*b or b*b+c*c==a*a:
    print("Tam giác vuông")
else:
    print("Tam giác thường")
```

```python
#B4
a,b = map(int, input().split())
if a==0 and b==0:
    print("Vô số nghiệm")
elif a==0 and b!=0:
    print("Vô nghiệm")
else:
    print(round(-b/a,2))
```

```python
#B5
a,b,c = map(int, input().split())
delta = float(b^2 - 4*a*c)
if delta < 0:
    print("Vô nghiệm")
elif delta == 0:
    print((-b)/(2*a))
else:
    print((-b+(delta)**(0.5))/(2*a), (-b-(delta)**(0.5))/(2*a))
```
