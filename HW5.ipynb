```python
#W6A16
n,m = map(int, input().split())
a = [[0 for j in range(m)] for i in range(n)]
for i in range(n):
    row = list(map(int, input().split()))
    for j in range(m):
        a[i][j] = row[j]
for i in range(n):
    for j in range(m-4):
        print(a[i][j], end=' ')
    print()
```

```python
#W6A17
n = int(input())
a = [[0 for j in range(n)] for i in range(n)]
for i in range(n):
    row = list(map(int, input().split()))
    for j in range(n):
        a[i][j] = row[j]
for i in range(n):
    print(a[i][i], end=' ')
print()
for i in range(n):
    print(a[i][n-i-1], end=' ')
```

```python
#W6A18
n,m,k = map(int, input().split())
srow = 0
scol = 0
a = [[0 for j in range(m)] for i in range(n)]
for i in range(n):
    row = list(map(int, input().split()))
    for j in range(m):
        a[i][j] = row[j]
for i in range(n):
    scol += a[i][k-1]
for j in range(m):
    srow += a[k-1][j]
print(scol, srow)
```

```python
#W6A23
a = [int(x) for x in input().split()]
k = int(input())
ptr = 0
while ptr < len(a):
    if a[ptr]<k:
        a.pop(ptr)
    else:
        ptr += 1
print(a)
```

```python
#W6A24
a = [int(x) for x in input().split()]
zeros = []
non_zeros = []
for num in a:
    if num == 0:
        zeros.append(num)
    else:
        non_zeros.append(num)
print(zeros + non_zeros)
```

```python
#W6A26
a = [int(x) for x in input().split()]
sm = 0
for i in range(1,len(a)-1):
    if a[i]>a[i-1] and a[i]>a[i+1]:
        sm += a[i]
print(sm)
```
